```python
import pandas as pd
import numpy as np
```

```python
attitudes = pd.read_csv('C:/Users/Jace/Documents/GitHub/MKT_Elections/data/attitudes.csv')
codebook = pd.read_csv('codebook.csv', header=None)
```

```python
attitudes[0:10]
codebook.columns = ['var_name', 'var_desc']
codebook[0:10]
```

```python
new_attitudes = attitudes[codebook['var_name']]
```

```python
new_attitudes[['VCF0004', 'VCF0901a']]
```

```python
new_attributes = new_attitudes.replace(' ', np.nan)
```

```python
mean_group = new_attitudes.groupby(['VCF0004', 'VCF0901a']).mean()
```

```python
new_attributes_mask = new_attributes.isnull().all()
new_attributes_mask
new_attributes_clean = new_attributes.dropna(axis=1,how='all')
new_attributes_clean.shape
```

```python
new_attributes_clean.dtypes
```

```python
new_attitudes_clean = new_attributes_clean.applymap(float)
```

```python
mean_group_clean = new_attitudes_clean.groupby(['VCF0004', 'VCF0901a']).mean()
mean_group_clean
```
